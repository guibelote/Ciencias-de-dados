# Redes Neurais e Deep Learning

Objetivos: rede densa simples para MNIST-like (usando tensorflow).

```
# Exemplo Keras (rede densa simples)
try:
    import tensorflow as tf
    from tensorflow import keras
    import numpy as np
    x = np.random.rand(100, 20)
    y = np.random.randint(0,2,size=(100,1))
    model = keras.Sequential([keras.layers.Dense(16, activation='relu', input_shape=(20,)), keras.layers.Dense(1, activation='sigmoid')])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(x,y, epochs=2, batch_size=16, verbose=0)
    print('Treino concluído')
except Exception as e:
    print('TensorFlow não disponível neste ambiente:', e)
```
