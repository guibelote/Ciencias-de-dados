# Big Data com Spark

Objetivos: exemplo WordCount com PySpark (local simulation).

```
# Exemplo WordCount com PySpark (se disponível)
try:
    from pyspark.sql import SparkSession
    spark = SparkSession.builder.master('local[*]').appName('wc').getOrCreate()
    data = spark.createDataFrame([('ola mundo','text'),('ola ola','text')], ['text','type'])
    words = data.rdd.flatMap(lambda r: r['text'].split()).map(lambda w: (w,1)).reduceByKey(lambda a,b: a+b)
    print(words.collect())
    spark.stop()
except Exception as e:
    print('PySpark não disponível:', e)
```
