# Fundamentos de Programação (Python)

Objetivos: variáveis, funções, condicionais, loops.

Projeto: Calculadora de notas e conversor de temperatura.

```
# Calculadora simples de médias e conversor de temperatura
def media_notas(notas):
    return sum(notas)/len(notas)

print("Média exemplo:", media_notas([7.5, 8.0, 9.0]))

def c_to_f(c): return (c * 9/5) + 32
print("25°C em °F:", c_to_f(25))
```
