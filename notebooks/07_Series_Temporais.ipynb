# Séries Temporais

Objetivos: manipular séries, decomposição, modelo ARIMA exemplo (simulado).

```
import pandas as pd
rng = pd.date_range('2020-01-01', periods=24, freq='M')
ts = pd.Series(range(24), index=rng)
print(ts.head())
# Decomposição (statsmodels may be required)
try:
    from statsmodels.tsa.seasonal import seasonal_decompose
    res = seasonal_decompose(ts, model='additive', period=12)
    res.plot()
except Exception as e:
    print("statsmodels não instalado no ambiente:", e)
```
